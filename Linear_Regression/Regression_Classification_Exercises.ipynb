```python
# Cell block 1
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read in data set from same file with pandas package
profitsDataSet = pd.read_csv("Rest_Pop_Profit.csv")

profitsDataSet = profitsDataSet.dropna()

# Reusable function for cleaning outlier data
# Parameters: Pandas Data Frame Object and Column being cleaned
def outlierDataSet(dataFrameObject,column):
        Q1 = dataFrameObject[column].quantile(0.25)                                                     # Data first quartile
        Q3 = dataFrameObject[column].quantile(0.75)                                                     # Data third quartile
        IQR = Q3 - Q1                                                                                   # Interquartile range calculated from first and third quartile
        lower_lim = Q1 - 1.5 * IQR                                                                      # Lower limit which will be assessing data
        upper_lim = Q3 + 1.5 * IQR                                                                      # Upper limit which will be assessing data
        outliers_15_low = dataFrameObject[column] < lower_lim                                           # Filtering for data below lower limit
        outliers_15_up = dataFrameObject[column] > upper_lim                                            # Filtering for data above upper limit
        outliersSubset = dataFrameObject[column][(outliers_15_low | outliers_15_up)]                    # Select all data outside of lower and upper limits
        return dataFrameObject.drop(outliersSubset.index)                                               # Return appended dataset with removed outliers

profitsDataSet = outlierDataSet(profitsDataSet,"Population")

# Allocate independent variable and dependent variable with data type conversion to list
independentVariable = np.array(profitsDataSet["Population"])
dependentVariable = np.array(profitsDataSet["Profits"])

# Reshape into 2-Dimensional numpy array
independentVariable = independentVariable.reshape(-1,1)

def regressionModelIntegerity(independentVariable,dependentVariable):
        X_train, X_test, Y_train, Y_test = train_test_split(independentVariable,dependentVariable,test_size=0.3,random_state=0) # Split data into individual variables for training and testing 
        model = LinearRegression().fit(X_train,Y_train)                                                                         # Create regression model
        y_train_prediction = model.predict(X_train)                                                                             # Train prediction with model prediction from x train
        y_test_pred = model.predict(X_test)                                                                                     # Predict the dependent variable from prediction with test data
        rmse = np.sqrt(mean_squared_error(Y_train,y_train_prediction))                                                          # Generate root mean square error from numpy
        r2 = r2_score(Y_train,y_train_prediction)                                                                               # Generate R^2 valuation from regression model
        print("Root mean squared error score:",rmse)                                                                            # Output root mean squared error score
        print("R^2 valuation:",r2)                                                                                              # Output R^2 valuation against test data

regressionModelIntegerity(independentVariable,dependentVariable)
```
